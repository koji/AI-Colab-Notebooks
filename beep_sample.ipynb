```python
import time
import jupyter_beeper as jb

pitch_freqs = {
    'G4': 392.00,
    'A4': 440.00,
    'C5': 523.25,
    'D5': 587.33,
    'E5': 659.25,
}

QUARTER_NOTE_DURATION = 0.8

song_data = [
    # 1小節目
    {'pitch': 'G4', 'value': 1.5},  # ソ (付点四分)
    {'pitch': 'G4', 'value': 0.5},  # ソ (八分)
    {'pitch': 'C5', 'value': 1.0},  # ド (四分)
    {'pitch': 'C5', 'value': 1.0},  # ド (四分)
    # 2小節目
    {'pitch': 'E5', 'value': 1.5},  # ミ (付点四分)
    {'pitch': 'D5', 'value': 0.5},  # レ (八分)
    {'pitch': 'C5', 'value': 1.0},  # ド (四分)
    {'pitch': 'E5', 'value': 1.0},  # ミ (四分)
    # 3小節目
    {'pitch': 'C5', 'value': 1.5},  # ド (付点四分)
    {'pitch': 'E5', 'value': 0.5},  # ミ (八分)
    {'pitch': 'G4', 'value': 2.0},  # ソ (二分)
    # 4小節目
    {'pitch': 'A4', 'value': 2.0},  # ラ (二分)
    {'pitch': 'A4', 'value': 2.0},  # ラ (二分)
    # 5小節目
    {'pitch': 'G4', 'value': 1.5},  # ソ (付点四分)
    {'pitch': 'G4', 'value': 0.5},  # ソ (八分)
    {'pitch': 'C5', 'value': 1.0},  # ド (四分)
    {'pitch': 'C5', 'value': 1.0},  # ド (四分)
    # 6小節目
    {'pitch': 'E5', 'value': 1.5},  # ミ (付点四分)
    {'pitch': 'D5', 'value': 0.5},  # レ (八分)
    {'pitch': 'C5', 'value': 1.0},  # ド (四分)
    {'pitch': 'A4', 'value': 1.0},  # ラ (四分)
    # 7小節目
    {'pitch': 'C5', 'value': 2.0},  # ド (二分)
    {'pitch': 'A4', 'value': 2.0},  # ラ (二分)
    # 8小節目
    {'pitch': 'G4', 'value': 3.0},  # ソ (付点二分)
    {'pitch': 'REST', 'value': 1.0}, # 四分休符
]

b = jb.Beeper()

print("start")

for note in song_data:
    duration = note['value'] * QUARTER_NOTE_DURATION
    if note['pitch'] == 'REST':
        print(f"休符: {duration:.2f}秒")
        time.sleep(duration)
    else:
        freq = pitch_freqs.get(note['pitch'])
        if freq:
            print(f"音: {note['pitch']} ({freq:.2f} Hz), 長さ: {duration:.2f}秒")
            b.beep(frequency=freq, secs=duration)
        else:
            print(f"未定義の音名: {note['pitch']}")
    time.sleep(0.5)

print("end")
```
