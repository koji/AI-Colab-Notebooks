```python
import torch
import soundfile as sf
from qwen_tts import Qwen3TTSModel
from IPython.display import Audio, display
```

```python
import torch
import soundfile as sf
from qwen_tts import Qwen3TTSModel

model = Qwen3TTSModel.from_pretrained(
    "Qwen/Qwen3-TTS-12Hz-1.7B-CustomVoice",
    device_map="cuda:0",
    dtype=torch.bfloat16,
    attn_implementation="flash_attention_2",
)
```

```python
# batch inference
wavs, sr = model.generate_custom_voice(
    text=[
        "She said she would be here by noon.",
        "彼女は正午までにはここに来ると言っていました。"
    ],
    language=["English", "Japanese"],
    speaker=["Ryan", "Ono_Anna"],
    instruct=["", "Very happy."]
)
sf.write("output_custom_voice_1.wav", wavs[0], sr)
sf.write("output_custom_voice_2.wav", wavs[1], sr)
```

```python
print("Playing output_custom_voice_1.wav:")
display(Audio("output_custom_voice_1.wav"))
```
